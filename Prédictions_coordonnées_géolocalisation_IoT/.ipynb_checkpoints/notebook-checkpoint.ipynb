<div align='justify'><h4><b><i>Contexte et objectifs</i></b></h4></div>

<div align='justify'>La géolocalisation Sigfox est un service de localisation à faible énergie et bas prix pour tous les appareils équipés du plus simple des modules Sigfox, sans GPS. Pour chaque message envoyé par un appareil, la géolocalisation Sigfox fournit ses coordonnées en temps réel (latitude/longitude) et la localisation précise en km, livré par une API standard. <br>
Dans ce contexte, <b>l’objectif du projet est d’utiliser les métadonnées des messages reçus par le réseau de stations Sigfox afin de prédire la localisation d’un appareil</b>. <br>
Un message est caractérisé par différentes informations : la base qui le reçoit, l’appareil qui l’émet, la puissance du signal (RSSI), son moment de réception. 
Dans cette étude, on dispose de 39250 messages, reçus par 259 bases et émis par 113 appareils. <br>
Dans un premier temps, la base de données d’entrainement fourni a été explorée afin de bien comprendre les interactions entre les bases et les devices. 
</div>
<div align='justify'><h4><b><i>Résumé de l'exploration de données</i></b></h4></div>
<div align='justify'>Après avoir constaté que <b>les données ne montrent pas de forte corrélation</b>, si ce n'est entre latitude et longitude, nous nous intéressons en détails à chacune des variables.<br>
<br>
Les <b>bases</b> sont au nombre de 259. A noter que si 160 sont présentes à la fois dans le jeu de test et dans le jeu d'entrainement, 8 ne sont présentes que dans le jeu de test.<br>
Les stations sont principalement situées aux Etats-Unis autour d'un axe allant du Wyoming au Texas en passant par le Colorado. <br>
Quelques stations sont des <b>outliers</b>, elles semblent mal localisées : elles reçoivent des messages avec un RSSI comparable à celui des autres stations, alors que localisées à 3000km au nord est du Canada !<br>
<br>
Les <b>devices</b> avec GPS (pour le jeu d'entrainement) sont au nombre de 113, localisés autour de l'axe cité précédemment.
Les devices sans GPS (jeu de test) sont au nombre de 56.<br>
<br>
39250 <b>messages</b> sont reçus par les bases dans le jeu d'entrainement, qui correspondent à 6068 messid différents. <br>
La plupart des messages sont par une (20%) deux (15%) ou 3 (14%) stations différentes, quelques uns par jusqu'à plusieurs dizaines et même jusqu'à 106 stations. <br>
<br>
Le <b>time_ux</b> nous indique que les messages ont été envoyés entre mai et décembre 2016, à des heures très variables essentiellement la journée mais aussi la nuit. <br>
La précision étant seulement de l'ordre de la seconde, un message a le même timestamp d'arrivée pour toutes les bornes qui le reçoivent : il n'est donc pas possible d'envisager une approche TDoA (time difference of arrival) pour ce jeu de données.<br>
Le rssi pourrait être sensible aux températures et montrer une saisonnalité, mais le jeu de données étant limité à 6 mois ne permet pas de l'étudier.<br>
<br>
Le <b>nseq</b> est une variable catégorielle (5 valeurs possibles). Ce paramètre semble être lié au mode de répétition d'envoi des messages d'après la documentation de l'API de sigfox, mais les valeurs non entières semblent plutôt indiquer que le jeu de données comprend déjà une moyenne ou une synthèse des répétitions ? Nous n'avons pas été capables d'interpréter ce champ, ni de le corréler aux autres variables.</div>

<div align='justify'><h4><b><i>Méthodologie - Modèle & Prédictions</i></b></h4></div>

<div align='justify'>Afin de réaliser nos prédictions sur les latitudes et longitudes des messages en fonction de la position des différentes balises nous avons procédé selon les étapes suivantes : <br><br>
<b>1. Clustering </b><br>
Dans un premier temps, nous avons réalisé un clustering des balise à la fois dans les datasets de train et de test à partir de leurs latitudes et leurs longitudes en utilisant la méthode des K-means (5 clusters réalisés).<br><br>
<b>2. Préparation des données</b><br>
<br>
<b> Traitement des outliers</b> <br>
A partir des données clusterisées, nous avons décidé d’éliminer les outliers en retirant les bases dont la longitude était inférieure à -80 et la latitude inférieure à 60 - Comme indiqué plus haut, nous avions identifié des bases largement excentrées par rapport aux autres.  <br><br>
<b> Création de la matrice de features</b><br>
Nous avons tout d’abord enrichi le dataset de nouvelles features: <br>
<li>La distance de la base à la moyenne de son cluster
<li>La distance de la base à la moyenne des bases 
<li>La latitude et la longitude converties en radians
<br>
<br>
Dans le dataset initial, chaque ligne correspondait à la réception d’un message par une base. Cela signifiait donc qu’un même message pouvait être répété plusieurs fois s’il était reçu par plusieurs balises. Pour chaque ligne, nous avions donc les coordonnées des balises propres à chacune d’entre elles avec les coordonnées du message en label, qui se répétait aussi autant de fois que le message était reçu par lesdites balises. Par conséquent, pour construire notre modèle, nous avions besoin de créer un dataset où chaque ligne représenterait un message unique, et trouver un moyen de condenser les informations relatives aux coordonnés des bases. Nous avons exploré deux options : <br>
<br>
1) Une ligne par message avec en colonnes l’intégralité des coordonnées exactes des balises et rssi associés<br>
2) Une ligne par message avec une moyenne des latitudes et longitudes des bases ayant reçu ledit message<br><br>
Après plusieurs essais, nous avons conclu que la meilleure performance était obtenue avec la 2ème option. Après des recherches, nous avons également rajouté à ces features en colonne le rssi pour chacune des bases prenant la valeur 0 si le message n’est pas reçu par la balise, et la valeur exacte du rssi si le message avait été reçu par la balise. <br>
Nous avons donc créé une matrice de features et une matrice « ground truth » (labels) sur ce modèle.<br>
<br>
<b>3. Sélection préliminaire de modèle</b><br>
Dans un premier temps, nous avons essayé trois modèles simples différents sur des datasets d’entraînements et de test : RandomForestRegressor, ExtraTreesRegressor et XGBoost. Pour chacun de ces modèles nous avons retenu comme score l’erreur cumulée à 80%. Avec des scores très proches, nous avons néanmoins retenu après plusieurs essais le modèle ExtraTreesRegressors qui a obtenu un score de 2954 vs. 2969 ou 2975.<br>
<br>
<b>4. Validation croisée - Leave One Out </b><br>
Pour réaliser la validation croisée, nous avons retenu la fonction LeaveOneGroupOut qui permet de réaliser une partition du dataset d’entraînement en laissant de côté un groupe d’échantillons de l’entraînement, et tester ainsi la robustesse du modèle en évitant l’overfitting. Comme précisé ci-dessus, nous avons utilisé le modèle ExtraTreesRegressor(n_estimators=200, max_features=0.7, max_depth=80, n_jobs=-1) dont nous avons cherché à optimiser les paramètres en testant plusieurs valeurs afin d'améliorer notre score. <br>
Après cross-validation, nous avons alors obtenu une erreur de distance cumulée à 80% moyennée de 4627m (vs. 7521m dans le modèle initial du cours).<br>
<br>

```python
# Définition de fonctions
def vincenty_vec(vec_coord):
    vin_vec_dist = np.zeros(vec_coord.shape[0])
    if vec_coord.shape[1] !=  4:
        print('ERROR: Bad number of columns (shall be = 4)')
    else:
        vin_vec_dist = [vincenty(vec_coord[m,0:2],vec_coord[m,2:]).meters for m in range(vec_coord.shape[0])]
    return vin_vec_dist

# evaluate distance error for each predicted point
def Eval_geoloc(y_train_lat , y_train_lng, y_pred_lat, y_pred_lng):
    vec_coord = np.array([y_train_lat , y_train_lng, y_pred_lat, y_pred_lng])
    err_vec = vincenty_vec(np.transpose(vec_coord))
    
    return err_vec
```

####<b><i>Création de nouvelles variables</i></b>


> <i>Passage des bases en colonnes avec comme valeurs RSSI, LAT, LNG</i>



```python
# Fonction pour passer les valeurs des bases en colonnes et y mettre la valeur du RSSI en log
def get_rssi(df_mess_train, listOfBs):

    df_mess_bs_group = df_mess_train.groupby(['messid'], as_index=False) # group data by message (messid)
    nb_mess = len(np.unique(df_mess_train['messid']))
    df_feat = pd.DataFrame(np.zeros((nb_mess,len(listOfBs))), columns = listOfBs) # feature matrix
    idx = 0
    id_list = [0] * nb_mess

    for key, elmt in df_mess_bs_group:
        df_mess_bs_group.get_group(key)
        #df_feat.loc[idx,df_mess_bs_group.get_group(key)['bsid']] = 1
        df_feat.loc[idx,df_mess_bs_group.get_group(key)['bsid']] = np.log(np.abs(df_mess_train.rssi))
        id_list[idx] = key
        idx = idx + 1  
    
    return df_feat, id_list

# Fonction pour passer les valeurs des bases en colonnes et y mettre la valeur de la latitude de la base
def get_lat(df_mess_train, listOfBs):

    df_mess_bs_group = df_mess_train.groupby(['messid'], as_index=False) # group data by message (messid)
    nb_mess = len(np.unique(df_mess_train['messid']))
    df_feat = pd.DataFrame(np.zeros((nb_mess,len(listOfBs))), columns = listOfBs) # feature matrix
    idx = 0
    id_list = [0] * nb_mess

    for key, elmt in df_mess_bs_group:
        df_mess_bs_group.get_group(key)
        #df_feat.loc[idx,df_mess_bs_group.get_group(key)['bsid']] = 1
        df_feat.loc[idx,df_mess_bs_group.get_group(key)['bsid']] = df_mess_train.bs_lat
        id_list[idx] = key
        idx = idx + 1    
    return df_feat, id_list

# Fonction pour passer les valeurs des bases en colonnes et y mettre la valeur de la longitude de la base
def get_lng(df_mess_train, listOfBs):

    df_mess_bs_group = df_mess_train.groupby(['messid'], as_index=False) # group data by message (messid)
    nb_mess = len(np.unique(df_mess_train['messid']))
    df_feat = pd.DataFrame(np.zeros((nb_mess,len(listOfBs))), columns = listOfBs) # feature matrix
    idx = 0
    id_list = [0] * nb_mess

    for key, elmt in df_mess_bs_group:
        df_mess_bs_group.get_group(key)
        #df_feat.loc[idx,df_mess_bs_group.get_group(key)['bsid']] = 1
        df_feat.loc[idx,df_mess_bs_group.get_group(key)['bsid']] = df_mess_train.bs_lng
        id_list[idx] = key
        idx = idx + 1      
    return df_feat, id_list


# 3 Fonctions pour mettre en forme les dataframes en sorties des fonctions précédentes et renommer les colonnes ds bases qui portent jusque la l'ID de la base
def format_matrice_bases_rssi(df_feat, id_list):
    matrice = pd.DataFrame()
    matrice = pd.concat([df_feat, pd.DataFrame(id_list)], axis=1).rename(columns={0: "messid"})  
    matrice.set_index('messid', inplace=True) 
    liste_col = []
    for col in matrice.columns:
      col = 'bsid_rssi_' + str(col) 
      liste_col.append(col)
        
    matrice.columns = liste_col

    return matrice, liste_col
  
def format_matrice_bases_lat(df_feat, id_list):
    matrice = pd.DataFrame()
    matrice = pd.concat([df_feat, pd.DataFrame(id_list)], axis=1).rename(columns={0: "messid"})  
    matrice.set_index('messid', inplace=True) 
    liste_col = []
    for col in matrice.columns:
      col = 'bsid_lat_' + str(col) 
      liste_col.append(col)
        
    matrice.columns = liste_col

    return matrice, liste_col

def format_matrice_bases_lng(df_feat, id_list):
    matrice = pd.DataFrame()
    matrice = pd.concat([df_feat, pd.DataFrame(id_list)], axis=1).rename(columns={0: "messid"})  
    matrice.set_index('messid', inplace=True) 
    liste_col = []
    for col in matrice.columns:
      col = 'bsid_lng_' + str(col) 
      liste_col.append(col)
        
    matrice.columns = liste_col

    return matrice, liste_col
```

> <i>Application des fonctions Data set de train</i>

```python
#On va essayer avec toutes les valeurs de rssi, lat et lng (autre possibilité : on moyenne ces valeurs pour chaque message)
df_bases_rssi, id_list_rssi = get_rssi(train, listOfBs)
df_bases_lat, id_list_lat = get_lat(train, listOfBs)
df_bases_lng, id_list_lng = get_lng(train, listOfBs)
```

```python
#Mise en forme de sdataframes précédents
bases_train_rssi, liste_bases_train_rssi = format_matrice_bases_rssi(df_bases_rssi, id_list_rssi)
bases_train_lat, liste_bases_train_lat = format_matrice_bases_lat(df_bases_lat, id_list_lat)
bases_train_lng, liste_bases_train_lng = format_matrice_bases_rssi(df_bases_lng, id_list_lng)
```

```python
# On calcule les centroïdes de chaque cluster, pour calculer la distance d'une base de son centroïde
# Dataset de train
mean_train = pd.DataFrame()
mean_train['lat_mean_cluster'] = train.groupby('cluster').bs_lat.mean()
mean_train['lng_mean_cluster'] = train.groupby('cluster').bs_lng.mean()

# Dataset de test
mean_test = pd.DataFrame()
mean_test['lat_mean_cluster'] = test.groupby('cluster').bs_lat.mean()
mean_test['lng_mean_cluster'] = test.groupby('cluster').bs_lng.mean()

# Jointure entre les datasets initiaux et les distances calculées, pour le TRAIN et le TEST
train=train.join(mean_train, on='cluster')
test=test.join(mean_test, on='cluster')
```

```python
# Nouvelles variables
def add_features(df):
  df['bs_lat_rssi'] = df.rssi * df.bs_lat
  df['bs_lng_rssi'] = df.rssi * df.bs_lng
  df['bs_lat_mean'] = df.bs_lat.mean()
  df['bs_lng_mean'] = df.bs_lng.mean()
  df["bs_lat_rad"] = np.radians(df["bs_lat"])
  df["bs_lng_rad"] = np.radians(df["bs_lng"])
  df['bs_dist_to_mean'] = Eval_geoloc(df.bs_lat, df.bs_lng, df.bs_lat_mean, df.bs_lng_mean)
  df['bs_dist_to_mean_cluster'] = Eval_geoloc(df.bs_lat, df.bs_lng, df.lat_mean_cluster, df.lng_mean_cluster)
  #df['dist_msg_base'] = Eval_geoloc(df.bs_lat, df.bs_lng, df.lat, df.lng)
```

```python
# Application de la fonction d'ajout de nouvelles features au dataframe de TRAIN
add_features(train)
```
